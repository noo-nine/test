```python
Heart = pd.read_csv('AHD.csv')
Heart.head()
```

```python
Heart.drop( columns = 'Unnamed: 0', inplace = True)
Heart
```

```python
Heart.shape
```

```python
# to know different values of ChestPain present in the column
Heart['ChestPain'].value_counts()
```

```python
Heart['Thal'].value_counts()
```

```python
# converting string values to numeric values
Heart['ChestPain'] = Heart['ChestPain'].map({'typical':0, 'nontypical':1, 'nonanginal':2, 'asymptomatic':3})
Heart['Thal'] = Heart['Thal'].map({'fixed':0, 'normal':1, 'reversable':2})
Heart['AHD'] = Heart['AHD'].map({'No':0, 'Yes':1})
```

```python
Heart.head()
```

```python
Heart.isnull().sum()
```

```python
Heart.dropna(inplace = True)
```

```python
Heart.isnull().sum()
```

```python
# all the other columns -- input variables
# AHD colum -- target variable
# separating X and y
X = Heart.drop(columns = 'AHD')
X
```

```python
y = Heart['AHD']
y
```

```python
# dividing it into training and testing datasets
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 21)
# test_size indicates the percent of data left for testing the model
# 70% is for training the model, 30% is for testing the model
```

```python
X_train
```

```python
X_test
```

```python
# some data has high numeric values, while some are small
# we can scale the data into same range
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# fit_transform does fit(X_train) and calculates their scaling parameters based on X_train data
# fit_transform also performs transform(X_train) to apply transformation to X_train
# for testing data we perform only transform() and not fit()
# sincve we want to use the scaling parameters learned from training data
# and apply the transformation only to the X_test
# otherwise there would be data leakage
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
# importimg the model
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state = 0).fit(X_train_scaled, y_train)
```

```python
# the model is now trained on X_train
```

```python
log_reg.predict(X_train_scaled)
```

```python
# to know how accurate these values are
log_reg.score(X_train_scaled, y_train)
```

```python
log_reg.predict(X_test_scaled)
```

```python
# on the testing data
log_reg.score(X_test_scaled, y_test)
```

```python
# accuracy for training data and testing data is almost same
# so our model is performing pretty good
```

```python
# now we try to make the model perform better by using some more parameters
```

```python
log_reg1 = LogisticRegression(random_state = 0, C = 1, fit_intercept = True).fit(X_train_scaled, y_train)
# C is a regularization parameter
```

```python
log_reg1.score(X_train_scaled, y_train)
```

```python
log_reg1.score(X_test_scaled, y_test)
```

```python
# we can see that the accuracy didnt improve much
# we can try chaning the parameters to make the model perform better
# the default vale of C is 1
# we can try adding C, fit_intercept models to any data
# the model may or may not perform better after doing the changes
# since model for every data will perform differently
```
