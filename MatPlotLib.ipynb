# Simple **Plot**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,5,4,2])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.title('Squares Plot')
plt.xlabel('number')
plt.ylabel('square')
plt.show()
```

# **Changing Original Apperance of The Figure**

```python
plt.figure(figsize=(4,4))
plt.plot([1,2,3,4], [1,4,9,16])
plt.axis([0, 5, 0,20])
plt.annotate('two square', (3,6))
plt.show()
```

# Changing into dots and their colour

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.show()
```

Bar Chart

```python
plt.clf()
# clears everything in the figure
```

```python
x = np.arange(4)
y = [6,10,4,8]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [6,10,4,8]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [6,10,4,8]
plt.xticks(x, ['A', 'B', 'C', 'D'])
plt.yticks(y, ['', 'High','', ''])
plt.bar(x,y)
plt.show()
```

```python
#adding colors
x = [1,2,3,4]
y = [6,10,4,8]
plt.bar(x, y, color = 'y')
plt.show()
```

```python
x = [1,2,3,4]
y = [6,10,4,8]
plt.bar(x, y, color = ['lime', 'r', 'lime', 'r'])
plt.show()
```

# **Plotting Two Sets Of Random Numbers**

```python
d = {'Red 0' : np.random.rand(10),
     'Grn S' : np.random.rand(10)}
df = pd.DataFrame(d)
df.plot(style = ['ro', 'gs'])
plt.show()
```

# **Time Series**

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods = 1000))
ts = ts.cumsum()
ts.plot()
plt.show()
```

# **Random Dots for a Scatter**

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
#varying sizes
x = np.random.rand(50)
y = np.random.rand(50)
area = np.pi * (np.random.rand(50))** 2
plt.scatter(x,y, s = area)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
area = np.pi * (20 * np.random.rand(50)) ** 2
plt.scatter(x, y, s = area)
plt.show()
```

```python
# adding colors
x = np.random.rand(50)
y = np.random.rand(50)
area = np.pi * (20 * np.random.rand(50)) ** 2
colors = np.random.rand(50)
plt.scatter(x , y, s = area, c = colors)
plt.show()
```
